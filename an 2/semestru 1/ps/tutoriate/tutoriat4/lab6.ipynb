```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# EXERCITIUL 1

```python
prob = 0.09
k = 7

def geom(p, n):
    return (1 - p) ** (n - 1) * p # P(Geom(p)=n)

print(f"P(k=7) = {geom(prob, k)}")

def density_geom(p, maxim):
    density = np.zeros(maxim + 1)
    for i in range(1, maxim + 1):
        density[i] = geom(p, i) # density[i] = P(Geom(p)=i)
    plt.style.use('dark_background')
    plt.stem(density)
    plt.xticks([i for i in range(1, maxim, maxim // 20)])
    plt.show()

def density_sim(p, n):
    results = np.zeros(n)
    for i in range(n):
        s = np.random.rand()
        days = 1 + np.floor(np.log(1-s)/np.log(1-p))
        results[i] = days
    l = [i-0.5 for i in range(int(np.floor(25/p)))]
    plt.style.use('dark_background')
    plt.hist(results, bins=l)
    plt.xticks([i for i in range(1, int(np.floor(25/p)), 20)])
    plt.yscale("log")
    plt.show()

density_geom(prob, 100)
density_sim(prob, 10000)
```

# EXERCITIUL 2

```python
avg = 20
target = 30

def pois(l, k):
    return (l ** k) * (np.e ** (-l)) / math.factorial(k)

def bin(p, n, k):
    return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

print(f"P(X=30) = {pois(avg, target)}")
print(f"Approximation: {bin(avg / 3600, 3600, 30)}")

def density_bin(p, n):
    results = np.zeros(n)
    for i in range(n):
        msg = 0
        for j in range(3600):
            s = np.random.rand()
            if s < p:
                msg += 1
        results[i] = msg
    plt.style.use('dark_background')
    plt.hist(results)
    plt.xticks([i for i in range(1, 40, 2)])
    plt.xlim((0, 40))
    plt.show()

def density_pois(l, maxim):
    density = np.zeros(maxim + 1)
    for i in range(1, maxim + 1):
        density[i] = pois(l, i)
    plt.style.use('dark_background')
    plt.stem(density)
    plt.xticks([i for i in range(1, maxim, maxim // 20)])
    plt.show()

density_bin(avg / 3600, 3600)
density_pois(20, 40)
```
